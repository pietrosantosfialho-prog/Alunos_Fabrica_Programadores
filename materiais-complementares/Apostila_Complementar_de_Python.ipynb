#Usando Funções

Como funções funcionam, funções tem nome, podem ou não receberem valores para fazer algo e retornar ou não valores.

Um exemplo simples e que já estamos utilizando desde o Hello World, é o print. print é uma função que recebe um valor de texto, no caso do nosso primeiro programa "Hello world" e ela imprime um valor na tela do usuário.

Outro exemplo é a função max do Python, essa função recebe múltiplos valores e retorna o maior valor. max(1,5,2) retorna 5, por exemplo.

O retorno de uma função pode ser usado:
*   Dando um nome ao valor usando uma variável
*   Passando o valor diretamente para outra função

Por exemplo:

```
maior_valor = max(1,5,2)
print(maior_valor)
> 5

```
ou, usando o valor de max diretamente em outra função:


```
print(max(1, 5, 2))
> 5
```

# Identação

## O que é identação?

Indentação é a forma de organizar o código usando espaços ou tabulações no início das linhas para indicar que determinadas instruções pertencem a um mesmo bloco ou estrutura. Em muitas linguagens de programação, como Python, a indentação não é apenas uma questão de estilo – ela é fundamental para a correta interpretação do código.

### Por que a indentação é importante?

### Organização Visual:

A indentação ajuda a tornar o código mais legível, facilitando a identificação das estruturas de controle (como laços, condições e funções).

### A identação é feita apertando 1 vez a tecla TAB, algumas IDEs fazem automaticamente a identação.


### Definição de Blocos:

Em Python, por exemplo, a indentação define quais linhas de código fazem parte de uma determinada estrutura. Sem a indentação correta, o interpretador não consegue identificar o início e o fim de um bloco.

### Exemplo Prático:

Imagine que queremos criar uma condição que verifica se um número é positivo. Veja como a indentação é utilizada:

Execute o bloco de código:


````python
numero = 5

if numero > 0:
    print("O número é positivo")
    print("Esta linha também faz parte do bloco do if")

print("Esta linha está fora do bloco do if")
````

Agora execute esse outro bloco:

````python
numero = -1

if numero > 0:
    print("O número é positivo")
    print("Esta linha também faz parte do bloco do if")

print("Esta linha está fora do bloco do if")
````

O que aconteceu com as saídas?

Por que as saídas aconteceram dessa forma?

Como o Python soube as linhas que precisavam ser executadas?

# Importando módulos

No Python, os módulos são conjuntos de funções, variáveis e classes que ajudam a organizar e reutilizar código. Em vez de escrever tudo do zero, você pode aproveitar módulos já prontos, como o módulo math, que contém diversas funções e constantes matemáticas.

Importando um Módulo
Para usar um módulo, você precisa importá-lo. O comando básico para importar um módulo é:



```
import math
```

Depois de importar, todas as funcionalidades do módulo math ficam disponíveis e você deve usá-las com o prefixo math.

Por exemplo, para calcular a raiz quadrada de um número, você utiliza a função sqrt:



```
import math

resultado = math.sqrt(16)
print(resultado)  # Saída: 4.0
```

O módulo math já vem com o Python e por essa razão não é necessário baixar nada com o [pip](https://docs.python.org/pt-br/3.13/installing/index.html), instalador de pacotes do Python.

Uma função inclusa no math é a factorial(), essa função é capaz de pegar o fatorial de um número, que é ele multiplicado pelos seus antecessores até o 1.

Por exemplo:

5 fatorial = 5! = 5 * 4 * 3 * 2 * 1

em Python:



```
import math

math.factorial(5)
```
